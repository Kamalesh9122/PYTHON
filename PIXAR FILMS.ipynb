```python
def replace_outliers(d1,column):
    Q1=d1[column].quantile(0.25)
    Q3=d1[column].quantile(0.75)

    IQR=Q3-Q1

    upper_bound=Q3+1.5*IQR  
    lower_bound=Q1-1.5*IQR

    # capping-replacing all higher values with upper bound (75% percentile)
    d1[column]=np.where(d1[column]>upper_bound,upper_bound,d1[column])
    # flooring -replace all lower values with lower bound (25% percentile)
    d1[column]=np.where(d1[column]<lower_bound,lower_bound,d1[column])
```

```python
replace_outliers(d1,"run_time")
```
